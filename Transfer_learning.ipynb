```python
import os
folders=os.listdir("transfer_learn_img")

idx=[i for i in range(len(folders))]
labels=dict(zip(folders,idx));
  

```

```python
image_data=[]
label=[]

for f in folders:
  path=os.path.join('transfer_learn_img',f)
  for img in os.listdir(path):
    image=load_img(os.path.join(path,img),target_size=(224,224))
    image=img_to_array(image)
    image_data.append(image)
    label.append(labels[f])
```

```python
import random 
combine=(list(zip(image_data,label)))

random.shuffle(combine)

xtrain[:],ytrain[:]=zip.(*combine)


```

```python
import numpy as np
from keras.utils import np.utils
xtrain=np.array(xtrain)
ytrain=np.array(ytrain)


ytrain=np.utils.to_categorical(ytrain)
```

**model**


```python
from keras.applications.resnet50 import ResNet50
from keras.optimizers import Adam
from keras.layers import *
from keras.models import Model
import keras
```

```python
model=ResNet50(include_top=False,weights='imagenet',input_shape=(224,224,3))
```

```python
model.summary()
```

```python


gap=GlobalAveragePooling2D()(model.output)
fc1=Dense(256,activation='relu')(gap)
d1=Dropout(0.4)(fc1)
fc2=Dense(4,activation='softmax')(d1)


model_new=Model(inputs=model.input,outputs=fc2)
model_new.summary()
```

```python
print(len(model_new.layers))
```

```python
##fine turning.....

adam=Adam(lr=0.00003)


for ix in range(169):
  model_new.layers[ix].trainable=False

    #setting layers untrainable upto a certain layer


model_new.compile(loss='categorical_crossentropy',optimizer=adam,metrics=['accuracy'])
```
